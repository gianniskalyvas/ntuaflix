```python
import mysql.connector

# Enter your credentials
connection = mysql.connector.connect(
    host="",
    user="",
    password="",
    database=""
)

cursor = connection.cursor()
```

```python
import pandas as pd

names_tsv = 'truncated_name.basics.tsv'

names = pd.read_csv(names_tsv, sep='\t')

names.replace('\\N', None, inplace=True)
names.replace({pd.NA: None}, inplace=True)

names
```

```python
insert_query_template = "INSERT INTO name_basics (name_id, name, birth_year, death_year, img_url) VALUES (%s, %s, %s, %s, %s);"

for index, row in names.iterrows():
    nconst = row['nconst']
    primary_name = row['primaryName']
    birth_year = row['birthYear']
    death_year = row['deathYear']
    img_url = row['img_url_asset']

    cursor.execute(insert_query_template, (nconst, primary_name, birth_year, death_year, img_url))
    

connection.commit()
```

```python
insert_query_template = "INSERT INTO primaryProfession (name_basics_name_id , profession ) VALUES (%s, %s);"

for index, row in names.iterrows():
  nconst = row['nconst']
  primary_profession = row['primaryProfession']
  if primary_profession is not None:
    l = primary_profession.split(',')
   
    for value in l:
      cursor.execute(insert_query_template, (nconst, value))
   

connection.commit()
```
